```python
#what is learning : find w that minimizes the loss
#Gradient descent algorithm
x_data=[1,2,3]
y_data=[2,4,6]

w=3.0  #a random guess : random value

def y_pred(x):
    return x * w

#loss function
def loss(x,y):
    return (y_pred(x)-y)*(y_pred(x)-y)

#compute gradient
def gradient(x,y):
    return 2*x*(x*w-y)

#before training

print("predict (before training)",4 ,y_pred(4))

#training loop
loss_list=[]
epoch_list=[]
w_list=[]
for epoch in range(20):
    for x_val,y_val in zip(x_data,y_data):
        grad=gradient(x_val,y_val)
        w=w-0.01*grad
        print("\tgrad:",x_val,y_val,grad)
        l=loss(x_val,y_val)
        
    print("progress:",epoch,"w=",w,"loss=",l)
    epoch_list.append(epoch)
    loss_list.append(l)
    w_list.append(w)
#after training
print("predict (after training)","4 hours", y_pred(4))

import matplotlib.pyplot as plt
plt.plot(epoch_list,loss_list,'b-',epoch_list,w_list,'r-')
plt.ylabel('loss & w')
plt.xlabel('epoch')
plt.show()
```

```python
#exercise 3-1  compute gradient
#Gradient descent algorithm
#suppose that y=3*x^2+2*x+6 [w2=3 w1=2 b=6] ,there are three patameters
x_data=[1,2,3]
y_data=[11,26,39]

w2=7.0  #a random guess : random value
w1=4.0
b=12

def y_pred(x):
    return x*x*w2+x*w1+b

#loss function
def loss(x,y):
    return (y_pred(x)-y)*(y_pred(x)-y)

#compute gradient respectively
def gradient_w2(x,y):
    return (-2*x*x*y)+2*w2*x*x*x*x+2*w1*x*x*x+2*b*x*x

def gradient_w1(x,y):
    return (-2*x*y)+2*w2*x*x*x+2*w1*x*x+2*b*x

def gradient_b(x,y):
    return (-2*y)+2*w2*x*x+2*w1*x+2*b

#before training

print("predict (before training)",4 ,y_pred(4))

#training loop
loss_list=[]
epoch_list=[]
w2_list=[]
w1_list=[]
b_list=[]
for epoch in range(7):
    for x_val,y_val in zip(x_data,y_data):
        g_w2=gradient_w2(x_val,y_val)
        g_w1=gradient_w1(x_val,y_val)
        g_b=gradient_b(x_val,y_val)
        w2=w2-0.01*g_w2
        w1=w1-0.01*g_w1
        b=b-0.01*g_b
        print("\tgrad:x=",x_val,"y=",y_val,"w2,w1,b gradient=",g_w2,g_w1,g_b)
        l=loss(x_val,y_val)
        
    print("progress:",epoch,"w2,w1,b=",w2,w1,b,"loss=",l)
    epoch_list.append(epoch)
    loss_list.append(l)
    w2_list.append(w2)
    w1_list.append(w1)
    b_list.append(b)
#after training
print("predict (after training)","4 hours", y_pred(4))

import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.plot(epoch_list,loss_list,'b-')
plt.title("loss gradient")
plt.ylabel('loss')
plt.xlabel('epoch')

plt.subplot(2,2,2)
plt.plot(epoch_list,w2_list,'r--')
plt.title("w2 gradient")
plt.ylabel('w2')

plt.subplot(2,2,3)
plt.plot(epoch_list,w1_list,'g--')
plt.title("w1 gradient")
plt.ylabel('w1')

plt.subplot(2,2,4)
plt.plot(epoch_list,b_list,'k--')
plt.title("b gradient")
plt.ylabel('b')
plt.xlabel('epoch')

plt.show()
```
