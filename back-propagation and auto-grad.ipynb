```python
#back-propagation and  autograd
import torch
from torch.autograd import Variable

x_data=[1,2,3]
y_data=[2,4,6]

#any random value
w=Variable(torch.Tensor([1.0]), requires_grad=True)
```

```python
#model function
def y_pred(x):
    return x * w

#loss function
def loss(x,y):
    return (y_pred(x) - y)*(y_pred(x) - y)

#before training
print("predict (before training)",4,y_pred(4).data[0])

epoch_list=[]
l_list=[]
w_list=[]
#training loop
for epoch in range(100):
    for x_val,y_val in zip(x_data,y_data):
        l=loss(x_val,y_val)
        #backward
        l.backward()
        print("\tgrad:",x_val,y_val,w.grad.data[0])
        w.data = w.data - 0.01 * w.grad.data
        
        #manually zero the gradients after updating weights
        w.grad.data.zero_()
    print("process:",epoch,w.data[0],l.data[0])
    l_list.append(float(l.data[0]))
    w_list.append(float(w.data[0]))
    epoch_list.append(epoch)
    
import matplotlib.pyplot as plt
print(len(l_list),len(w_list),len(epoch_list))
plt.subplot(1,2,1)
plt.plot(epoch_list,l_list,'r-')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.subplot(1,2,2)
plt.plot(epoch_list,w_list,'k-')
plt.ylabel('w')
plt.xlabel('epoch')

plt.show()

#after training
print("predict (after training)", 4 ,y_pred(4).data[0])
```
