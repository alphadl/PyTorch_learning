```python
#logistic regression in pytorch
#binary prediction
'''
linear regression: x-->Linear-->y_pred

logistic regression: x-->Linear-->Sigmoid-->y_pred

Sigmoid(z)=1/(1+e^(-z))
>>z belongs [-inf,inf]
>>Sigmoid(z) belongs (0,1)   y_pred = 1 if y_pred > 0.5 else 0

loss function is Binary Cross 
'''
#load datasets
import torch
from torch.autograd import Variable
x_data=Variable(torch.Tensor([[1],[2],[3],[4]]))
y_data=Variable(torch.Tensor([[0],[0],[1],[1]]))


import torch.nn.functional as F

class Model(torch.nn.Module):
    
    def __init__(self):
        super(Model,self).__init__()
        self.linear=torch.nn.Linear(1,1)
    
    def forward(self,x):
        y_pred=F.sigmoid(self.linear(x))
        return y_pred

#our model
model=Model()

criterion = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)

#training loop
for epoch in range(1000):
    y_pred=model(x_data)
    
    #compute and print loss
    loss=criterion(y_pred,y_data)
    print(epoch,loss.data[0])
    
    #zero gradients ,perform a backward pass ,
    optimizer.zero_grad()
    loss.backward()
    
    #and update the weights
    optimizer.step()

#after training
hour_var=Variable(torch.Tensor([[1]]))
print("predict 1 hour ",1.0,model(hour_var).data[0][0] > 0.5)

hour_var=Variable(torch.Tensor([[7]]))
print("predict 7 hour ",7.0,model(hour_var).data[0][0] > 0.5)
```
